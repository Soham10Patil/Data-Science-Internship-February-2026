```python
#Problem Statement 1:Smart Parking Lot Management System
def parking_system(capacity, logs):
    current = 0
    peak = 0
    for log in logs:
        if log == "IN":
            current += 1
        elif log == "OUT":
            current -= 1
        if current > peak:
            peak = current
    if current > capacity:
        status = "Full"
    else:
        status = "Available"
    return current, peak, status
capacity = 50
logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]
current, peak, status = parking_system(capacity, logs)
print("Currently Parked Vehicles:", current)
print("Peak Usage:", peak)
print("Parking Status:", status)
```

```python
# Problem Statement 2:Online Food Delivery Time Estimator
def estimate_delivery_time(distance, traffic, weather):
    time = distance * 5
    if traffic == "High":
        time += 10
    elif traffic == "Medium":
        time += 5
    if weather == "Rainy":
        time += 5
    elif weather == "Storm":
        time += 15
    elif weather == "Cloudy":
        time += 5
    return time
distance = 8
traffic = "High"
weather = "Rainy"
eta = estimate_delivery_time(distance, traffic, weather)
print("Estimated Delivery Time:", eta, "minutes")
```

```python
#Problem Statement 3:Movie Theatre Seat Occupancy Analyzer
def seat_occupancy(total_seats, booked_seats):

    booked_count = len(booked_seats)

    occupancy = (booked_count / total_seats) * 100

    if occupancy == 100:
        status = "Housefull"
    elif occupancy >= 70:
        status = "Almost Full"
    else:
        status = "Seats Available"

    return occupancy, status


total = 200
booked = [1] * 150

percent, status = seat_occupancy(total, booked)

print(f"Occupancy: {percent:.0f}%")
print("Show Status:", status)
```

```python
#Problem Statement 4:Cloud Server Load Classification System
def classify_server_load(cpu_readings):

    if isinstance(cpu_readings, int):
        cpu_readings = [cpu_readings]

    avg = sum(cpu_readings) / len(cpu_readings)

    if avg < 50:
        status = "Normal"
    elif avg <= 80:
        status = "Warning"
    else:
        status = "Critical"

    return round(avg), status
avg, status = classify_server_load([45, 60, 70, 85, 90])
print("Average CPU Load:", avg)
print("Server Load Status:", status)
```

```python
# Problem Statement 5:Smart Classroom Resource Usage Monitor
def monitor_resources(usage):
    limits = {
        "Projector": 5,
        "AC": 8,
        "Lights": 6
    }

    overused = []
    for resource in usage:
        if usage[resource] > limits[resource]:
            overused.append(resource)
    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")
monitor_resources({"Projector": 4, "AC": 7, "Lights": 5})
```

```python
# Problem Statement 6: Online Event Registration Capacity Controller
def manage_registration(capacity, registrations):
    confirmed = 0
    waitlisted = 0
    for _ in range(registrations):
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlisted += 1
    if confirmed >= capacity:
        status = "Closed"
    else:
        status = "Open"
    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)
manage_registration(100, 105)
```
