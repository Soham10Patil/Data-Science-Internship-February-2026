```python
#Problem 1: Mobile Recharge Validation System
valid_plan = [199, 299, 399, 599]
def valid(amount):
  if amount < 50:
    print("Minimum recharhe should be 50")
    return False
  if amount not in valid_plan:
    print("Invalid plan")
    return False
  return True
while True:
  amt = int(input("Enter the amount: "))
  if valid(amt):
    print("Recharge successful")
  else:
    print("Recharge failed")

```

```python
# Problem 2: Inventory Reorder Alert System
inventory = {"Milk" : 7 , "Peanuts(kg)" : 19, "butter" : 12}
def stock_check(inventory):
  for item, quantity in inventory.items():
    if quantity < 15:
      print(f"{item} => Low Stock - Reorder")
    else:
      print(f"{item} => Stock OK")
stock_check(inventory)
```

```python
# Problem 3: Student Result Processing System
def process_result(marks):
    total = 0
    for m in marks:
        total += m
    avg = total / len(marks)
    if avg >= 50:
        return avg, "Pass"
    else:
        return avg, "Fail"

marks = [60, 45, 70, 80]
average, result = process_result(marks)
print(f"Average: {average}")
print(f"Result: {result}")
```

```python
# Problem 4: Cab Fare Estimator with Retry Option
BASE_FARE = 50
RATE_PER_KM = 12


def calculate_fare(km, peak):
    fare = BASE_FARE + (RATE_PER_KM * km)

    if peak:
        fare *= 1.25   # add 25%

    return fare


while True:
    km = float(input("Enter distance (km): "))
    peak_input = input("Peak hour? (y/n): ").lower()

    peak = True if peak_input == 'y' else False

    fare = calculate_fare(km, peak)

    print(f"Total Fare: ₹{fare:.2f}")

    retry = input("Calculate again? (y/n): ").lower()
    if retry != 'y':
        break
```

```python
# Problem 5: Employee Attendance Eligibility Checker
def check_eligibility(attendance):
    present = 0

    # count present days
    for day in attendance:
        if day == "P":
            present += 1

    percentage = (present / len(attendance)) * 100

    if percentage >= 75:
        return percentage, "Eligible"
    else:
        return percentage, "Not Eligible"


attendance = ["P", "P", "A", "P", "P"]

percent, status = check_eligibility(attendance)

print(f"Attendance: {percent:.2f}%")
print(f"Status: {status}")
```

```python
# Problem 6: Password Strength Checker
def check_password(password):
    if len(password) < 8:
        return "Weak (Too short)"

    has_digit = False
    has_special = False
    special_chars = "@#$"

    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in special_chars:
            has_special = True

    if has_digit and has_special:
        return "Strong"
    else:
        return "Weak"


pwd = input("Enter password: ")
result = check_password(pwd)
print(result)

```
