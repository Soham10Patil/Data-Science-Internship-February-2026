# 1. Difference between "Love" and "love" in NLP
# => In NLP, "Love" and "love" are treated as different tokens if case normalization is not applied.
# This means the model may interpret them as separate words, which can reduce consistency and
# affect learning. Converting all text to lowercase ensures both are treated as the same word,
# improving accuracy and reducing vocabulary size.

# 2. What happens if stopwords are not removed?
# => If stopwords are not removed, the dataset will contain many common words like "is", "the", "and"
# that do not add much meaning. This increases noise, slows down processing, and can reduce model 
# performance because important words get less attention.

# 3. Two real-world scenarios where removing stopwords is harmful
# => Sentiment Analysis: Words like "not" are critical.
# Example: "I am not happy" → removing "not" changes meaning completely.
# Question Answering Systems: Words like "what", "where", "how" are essential for understanding intent.

# 4. Difference between Stemming and Lemmatization
# => Stemming: Cuts words to root form using simple rules (may not be meaningful)
# Example: "running" → "run", "better" → "bet"
# Lemmatization: Converts words to meaningful base form using vocabulary and grammar
# Example: "running" → "run", "better" → "good"
# Lemmatization is more accurate but slower than stemming.

```python
import re
from collections import Counter
#2
def preprocess_text(text):
    # Handle empty or None input
    if not text or not isinstance(text, str):
        return ""

    # Lowercase
    text = text.lower()

    # Remove URLs
    text = re.sub(r'http\S+|www\S+', '', text)

    # Remove emails
    text = re.sub(r'\S+@\S+', '', text)

    # Remove numbers
    text = re.sub(r'\d+', '', text)

    # Handle repeated characters
    text = re.sub(r'(.)\1{2,}', r'\1\1', text)

    # Remove emojis & special characters
    text = re.sub(r'[^\w\s]', '', text)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # Remove short words (≤2) except "no", "not"
    words = text.split()
    filtered_words = [
        word for word in words
        if len(word) > 2 or word in ["no", "not"]
    ]

    return " ".join(filtered_words)

```

```python
sample = "WOW!!! This is soooo goooood!!! I have 2 dogs. Email me at test@gmail.com"

print(preprocess_text(sample))
```

```python
#3
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]
```

```python
for text in test_sentences:
    cleaned = preprocess_text(text)
    tokens = cleaned.split()

    print("Original Text   :", text)
    print("Cleaned Tokens  :", tokens)
    print("Cleaned Sentence:", cleaned)
    print("-" * 50)
```

```python
#5
def token_analytics(text):
    cleaned = preprocess_text(text)
    tokens = cleaned.split()

    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))

    if total_tokens > 0:
        avg_token_length = sum(len(word) for word in tokens) / total_tokens
    else:
        avg_token_length = 0

    return {
        "original": text,
        "cleaned": cleaned,
        "tokens": tokens,
        "total_tokens": total_tokens,
        "unique_tokens": unique_tokens,
        "avg_token_length": round(avg_token_length, 2)
    }
```

```python
results = []

for text in test_sentences:
    analysis = token_analytics(text)
    results.append(analysis)

    print("Original Text        :", analysis["original"])
    print("Cleaned Sentence     :", analysis["cleaned"])
    print("Tokens               :", analysis["tokens"])
    print("Total Tokens         :", analysis["total_tokens"])
    print("Unique Tokens        :", analysis["unique_tokens"])
    print("Average Token Length :", analysis["avg_token_length"])
    print("-" * 60)
```

```python

all_tokens = []

for text in test_sentences:
    cleaned = preprocess_text(text)
    tokens = cleaned.split()
    all_tokens.extend(tokens)

word_freq = Counter(all_tokens)

top_10 = word_freq.most_common(10)

print("Top 10 Most Frequent Words:")
for word, count in top_10:
    print(f"{word}: {count}")

least_5 = sorted(word_freq.items(), key=lambda x: x[1])[:5]

print("\nTop 5 Least Frequent Words:")
for word, count in least_5:
    print(f"{word}: {count}")
```

```python
#6
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        cleaned = preprocess_text(text)

        # Skip completely empty outputs
        if not cleaned:
            continue

        tokens = cleaned.split()

        clean_sentences.append(cleaned)
        all_tokens.extend(tokens)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences,
        "total_sentences": len(clean_sentences),
        "total_tokens": len(all_tokens)
    }


result = full_pipeline(test_sentences)

print("Clean Sentences:")
for sent in result["clean_sentences"]:
    print("-", sent)

print("\nAll Tokens:")
print(result["tokens"])
```

```python
#7
edge_cases = [
    "",                     # Empty string
    "😂😂😂😂😂",             # Only emojis
    "1234567890",           # Only numbers
]

result = full_pipeline(edge_cases)

print(result)
```
