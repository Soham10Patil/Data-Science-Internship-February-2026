```python
# 2. Pass / Fail Analyzer
marks = [45, 78, 90, 33, 60]
a = 0
b = 0
for i in marks:
  if i > 50:
    print(f"{i} -> Pass")
    a += 1
  else:
    print(f"{i} -> Fail")
    b += 1
print(f"pass => {a}")
print(f"fail => {b}")
```

```python
# 3. Simple Data Cleaner
from os.path import split
names = [" Alice ", "bob", " CHARLIE "]
exspaces = [name.strip() for name in names]
print(f"clean(extra spaces removed) => {exspaces}")
lowercase = [name.lower() for name in exspaces]
print(f"clean list => {lowercase}")
```

```python
# 4. Message Length Analyzer
messages = ["Hi", "Welcome to the platform", "OK"]
lenMessage = [len(msg) for msg in messages]
for i in range(len(messages)):
  if lenMessage[i] > 10 :
    print(f"{messages[i]} => {lenMessage[i]} (Flaged msg more that 10 char)")
  else:
     print(f"{messages[i]} => {lenMessage[i]}")
```

```python
# 5. Error Message Detector
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
c=0
for i in range(len(logs)):
  if logs[i] == "ERROR":
    c = c + 1
print(f"Total ERROR count => {c}")
```
