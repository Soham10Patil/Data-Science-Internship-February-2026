```python
# Problem 1: Social Media – Post Engagement Analyzer
def analyze_engn(likes):
  total = 0
  for i in likes:
    total = total + i
  print(f"Total : {total}")
  if total >= 1000:
    print(f"Viral post")
  else:
    print(f"Normal engagement")

like = [543, 743, 894, 732]
analyze_engn(like)
```

```python
# Problem 3: Agriculture – Rainfall Adequacy Checker
def rainfall_status(rainfall, required_level):
    total = 0

    # loop to calculate sum
    for r in rainfall:
        total += r

    avg = total / len(rainfall)

    if avg >= required_level:
      status = "Adequate Rainfall"
    else:
      status = "Inadequate Rainfall"

    return avg, status


# Example usage
rainfall = [60, 75, 80, 70, 65, 82]
required_level = 70

average, result = rainfall_status(rainfall, required_level)

print("Average Rainfall:", average)
print("Rainfall Status:", result)
```

```python
# Problem 4: Social Media – Duplicate Account Detection
def dup_account(usernames):
    unique_users = set(usernames)
    if len(unique_users) < len(usernames):
        return "Yes"
    else:
        return "No"
username = ["soham", "sanjana", "piyush", "ravi", "piyush"]
result = dup_account(username)
print(f"Duplicate Accounts Found: {result}")
```

```python
# Problem 6: Agriculture – Premium Crop Price Filter
def premium_crops(prices):
    premium = []
    for price in prices:
        if price > 2000:
            premium.append(price)

    return premium
prices = [1500, 2500, 1800, 3200]
result = premium_crops(prices)
print("Premium Crops:", result)
```

```python
# Problem 7: System Monitoring – Application Health
def check_health(errors):
    if errors == 0:
        status = "Healthy"
    elif errors <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    return status

error_count = 7
result = check_health(error_count)
print("Error Count:", error_count)
print("System Status:", result)
```

```python
# Problem 8: Banking – Daily Transaction Limit Checker
def limit_checker(amount, daily_limit):
  if amount <= daily_limit:
    return "Within Daily Limit : Approved"
  else:
    return "Exceeds Daily Limit : Rejected"
amount = int(input("Enter amount for transaction : "))
daily_limit = 50000
result = limit_checker(amount, daily_limit)
print(f"Transaction Status : {result}")
```

```python
# Problem 9: E-Learning – Student Attendance Eligibility System
def attendence_eli(attendence):
  total = 0
  for i in attendence:
    if i == "P":
      total = total + 1
  percentage = (total / len(attendence)) * 100
  if percentage >= 75:
    status = "Eligible"
  else:
    status = "Not Eligible"
  return percentage, status

attendence = ['P', 'P', 'A', 'P', 'A']
percentage, result = attendence_eli(attendence)
print(f"Attendence Percentage : {percentage}")
print(f"Student Status : {result}")
```

```python
# Problem 10: Smart Electricity Bill Analyzer
def electicity_bill(unit):
  if unit <= 100:
    bill = unit * 3
  elif unit > 100 and unit <= 200:
    bill = unit * 5
  else:
    bill = unit * 7

  if bill < 500:
    status = "low usage"
  elif bill > 500 and bill <= 1000:
    status = "modeate usage"
  else:
    status = "high usage"
  return bill, status
unit = int(input("Enter units of electricity consumed : "))
bill, result = electicity_bill(unit)
print(f"Electricity Bill : {bill}")
print(f"Usage : {result}")
```
