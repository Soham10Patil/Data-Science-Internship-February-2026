```python
#Problem 1: Smart Light Controller
class light_controller:
    def __init__(self,room):
        self.room = room
        self.status = "OFF"
    def turn_on(self):
        self.status = "ON"
    def turn_off(self):
        self.status = "OFF"
    def display(self):
        print(f"{self.room} is {self.status}")
light = light_controller("Bedroom")
action = input(f"Light(ON/OFF)")
if action == "ON":
    light.turn_on()
elif action == "OFF":
    light.turn_off()
light.display()
```

```python
# Problem 2: Employee ID Card System
class id_card_sys:
  def __init__(self, id, name, dept):
    self.id = id
    self.name = name
    self.dept = dept
  def display (self):
    print(f"ID: {self.id}\nName: {self.name}\nDepartment: {self.dept}")
ID = input("Enter ID: ")
Name = input("Enter Name: ")
Dept = input("Enter Department: ")
obj = id_card_sys(ID, Name, Dept)
obj.display()
```

```python
# Problem 3: Mobile Contact Record
class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone
    def save_contact(self):
        print("Contact Saved")
    def display_contact(self):
        print("Name:", self.name)
        print("Phone:", self.phone)
name = input("Contact Name: ")
phone = input("Phone Number: ")
c1 = Contact(name, phone)
c1.save_contact()
c1.display_contact()
```

```python
#Problem 4: Product Price Tag Generator
class tag_generator:
  def __init__(self, product, price):
    self.product = product
    self.price = price
  def display(self):
    print(f'product: {self.product}')
    print(f'price: {self.price}')
prod = input("Product: ")
price = input("Price: ")
obj = tag_generator(prod, price)
obj.display()
```

```python
#Problem 5: Movie Rating Display System
class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating
    def display(self):
        print(f"Movie: {self.name}")
        print(f"Rating: {self.rating} / 5")
name = input("Movie Name: ")
rating = float(input("Rating: "))
m1 = Movie(name, rating)
m1.display()
```

```python
# Problem 6: Delivery Address Manager
class Delivery:
    def __init__(self, name, address):
        self.name = name
        self.address = address
    def print_details(self):
        print("Delivery Details")
        print(f"Customer: {self.name}")
        print(f"Address: {self.address}")
name = input("Customer Name: ")
address = input("Address: ")
d1 = Delivery(name, address)
d1.print_details()
```
